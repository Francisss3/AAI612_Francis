```python
from newspaper import Article
import pandas as pd

urls = [
    "https://news.northeastern.edu/2023/12/01/military-ai/",
    "https://www.yahoo.com/news/century-long-reign-machine-gun-070601347.html",
    "https://time.com/7202584/gaza-ukraine-ai-warfare/",
    "https://apnews.com/article/israel-palestinians-ai-technology-737bc17af7b03e98c29cec4e15d0f108",
    "https://www.972mag.com/lavender-ai-israeli-army-gaza/",
    "https://www.aljazeera.com/news/2024/4/4/ai-assisted-genocide-israel-reportedly-used-database-for-gaza-kill-lists",
    "https://www.npr.org/2023/12/14/1218643254/israel-is-using-an-ai-system-to-find-targets-in-gaza-experts-say-its-just-the-st",
    "https://www.democracynow.org/2024/4/5/israel_ai",
    "https://impactpolicies.org/news/359/occupied-palestine-ai-and-indiscriminate-targeting",
    "https://truthout.org/articles/report-israeli-army-uses-ai-to-produce-palestinian-targets-for-assassination/",
    "https://www.bbc.com/news/articles/cwy3l02wxqdo",
    "https://apnews.com/article/israel-lebanon-hezbollah-pagers-hamas-826a71b87e395f0195f877e4e9d238f1",
    "https://www.justsecurity.org/108719/israel-pager-operation-hezbollah/",
    "https://edition.cnn.com/2024/09/19/middleeast/lebanon-pager-walkie-talkie-attacks-explainer-intl-hnk/index.html",
    "https://www.rusi.org/explore-our-research/publications/commentary/israels-targeting-ai-how-capable-it",
    "https://lieber.westpoint.edu/beyond-headlines-combat-deployment-military-ai-based-systems-idf/",
    "https://pahrw.org/en/اصدارات/تقارير-ودراسات/ai-plotted-geocide-how-corporations-facilitate-israels-ai-enabled-war-on-gaza/"
]

from newspaper import Article
import pandas as pd

articles = []

for url in urls:
    try:
        article = Article(url)
        article.download()
        article.parse()

        publish_date = article.publish_date
        title = article.title
        text = article.text.replace('\n', ' ')

        articles.append([publish_date, title, text])
        print(f"Successfully collected: {title}")

    except Exception as e:
        print(f"Failed to collect {url}: {e}")

df_articles = pd.DataFrame(articles, columns=['Date', 'Title', 'Text'])
df_articles.to_csv('ai_warfare_articles.csv', index=False)
```

```python
# Install libraries for preprocessing
!pip install nltk
import nltk
nltk.download('stopwords')

import pandas as pd
import re
import nltk
from nltk.corpus import stopwords

# Load your data
df_articles = pd.read_csv('ai_warfare_articles.csv')

# Define text cleaning function
def clean_text(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = text.split()
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return ' '.join(tokens)

# Apply cleaning
df_articles['Cleaned_Text'] = df_articles['Text'].apply(clean_text)

# Save and inspect the cleaned data
df_articles.to_csv('cleaned_ai_warfare_articles.csv', index=False)
df_articles.head()
```

```python
!pip install transformers torch

import pandas as pd
from transformers import pipeline

# Load cleaned dataset
df_articles = pd.read_csv('cleaned_ai_warfare_articles.csv')

# Initialize pretrained sentiment analysis pipeline
sentiment_analyzer = pipeline('sentiment-analysis')

# Apply sentiment analysis (BERT-based model)
df_articles['Sentiment'] = df_articles['Cleaned_Text'].apply(
    lambda x: sentiment_analyzer(x[:512])[0]['label']
)

# Save analyzed dataset
df_articles.to_csv('final_sentiment_results.csv', index=False)

# Check results
df_articles[['Title', 'Sentiment']].head()
```
